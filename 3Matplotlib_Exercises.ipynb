## Exercise 1

* **Create a figure object called fig using plt.figure() **
* **Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax. **
* **Plot (x,y) on that axes and set the labels and titles to match the plot below: **

## Exercise 2
**Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.**

```python
from matplotlib.ticker import MaxNLocator


fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1])
axes2 = fig.add_axes([0.2, 0.5, .2, .2])

axes2.xaxis.set_major_locator(MaxNLocator(5))
axes2.yaxis.set_major_locator(MaxNLocator(5))
```

**Now plot (x,y) on both axes. And call your figure object to show it.**

```python
from matplotlib.ticker import MaxNLocator
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1])
axes2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])
axes2.xaxis.set_major_locator(MaxNLocator(4))
axes2.yaxis.set_major_locator(MaxNLocator(4))

plt.plot(x,y,'r')

axes1.plot(x,y,'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_xlim([0, 100])
axes1.set_ylim([0, 200])



axes2.plot(x,y,'red')
axes2.set_xlabel('x')
axes2.set_ylabel('y')
axes2.set_xlim([0, 100])
axes2.set_ylim([0, 200])

plt.show()
```

## Exercise 4

**Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

**Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
from matplotlib.ticker import MaxNLocator


axes[0].plot(x,y,color="blue", lw=3, ls='--')
axes[0].set_xlim([0, 100])
axes[0].set_ylim([0, 200])
axes[0].yaxis.set_major_locator(MaxNLocator(4)) 

axes[1].set_xlim([0, 100])
axes[1].set_ylim([0, 10000])
axes[1].plot(x,z,color="red", lw=3, ls='-')
axes[1].yaxis.set_major_locator(MaxNLocator(5)) 


fig
```

*See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.*

```python
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(12,2))

axes[0].plot(x,y,'b', lw=5)
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_xlim([0, 100])
axes[0].set_ylim([0, 200])

axes[1].plot(x,z,'r', lw=3, ls='--')
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')
axes[1].set_xlim([0, 100])
axes[1].set_ylim([0, 10000])
```
